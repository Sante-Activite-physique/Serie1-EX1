```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Read the (text) data file into numpy array
data_file =(r'C:/Users/SCD UM/Desktop/Python/SNS/Exercice1/DAT/ECGu.txt') 

# Lecture du fichier
donnees = np.loadtxt(data_file)


# Extraction des colonnes
colonne1 = (donnees[:, 0]*0.520)/1000
colonne2 = (donnees[:, 1]*0.520)/1000
colonne3 = (donnees[:, 2]*0.520)/1000

# Création du graphique
time=np.arange(len(donnees))/1024
#temps = np.arange(len(colonne1)) / 1024 # Conversion en temps (échantillons / fréquence d'échantillonnage)
plt.plot(time, colonne1, label="Lead I")
plt.plot(time, colonne2, label="Lead II")
plt.plot(time, colonne3, label="Lead III")
plt.xlabel("Temps (s)")
plt.ylabel("Signal (µV)")
plt.title("Eléctrocardiogramme ")
plt.legend()
plt.xlim(0,7,0.5)
plt.ylim(-0.4,0.8,0.2)

# Affichage du graphique
plt.show()
```
